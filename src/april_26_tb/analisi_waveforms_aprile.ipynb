```python
import uproot
import awkward as ak

# =========================
# 1. APRI FILE
# =========================
file = uproot.open("~/Scrivania/sampic_run51_merged.root")
tree = file["picoTree"]

# =========================
# 2. CONTENITORI OUTPUT
# =========================
det_out = []
multiplicity_out = []
cell0_out = []
array_size_out = []
x_out = []
y_out = []
wf_out = []

# =========================
# 3. LETTURA A CHUNK (FONDAMENTALE)
# =========================
for chunk in tree.iterate(
    ["Detector", "Multiplicity", "Cell0TimeStamp", "ArraySize", "xCoord", "yCoord", "Waveform"],
    step_size=5000,
    library="ak"
):

    # =========================
    # 4. FILTRO PER CHUNK
    # =========================
    mask= (chunk["xCoord"] == 5) & (chunk["yCoord"] == 5)
    filtered = chunk[mask]

    # =========================
    # 5. ACCUMULA RISULTATI
    # =========================
    det_out.append(filtered["Detector"])
    multiplicity_out.append(filtered["Multiplicity"])
    cell0_out.append(filtered["Cell0TimeStamp"])
    array_size_out.append(filtered["ArraySize"])
    x_out.append(filtered["xCoord"])
    y_out.append(filtered["yCoord"])
    wf_out.append(filtered["Waveform"])
```

```python
# =========================
# 6. CONCATENA TUTTO
# =========================

det_out = ak.concatenate(det_out)
multiplicity_out = ak.concatenate(multiplicity_out)
cell0_out = ak.concatenate(cell0_out)
array_size_out = ak.concatenate(array_size_out)
x_out = ak.concatenate(x_out)
y_out = ak.concatenate(y_out)
wf_out = ak.concatenate(wf_out)

# =========================
# 7. DEBUG
# =========================
print("Eventi selezionati:", len(x_out))

# =========================
# 8. SALVA FILE ROOT
# =========================
#with uproot.recreate("filtered.root") as f:
#    f["picoTree"] = {
#        "xCoord": x_out,
#       "yCoord": y_out,
#        "Waveform": wf_out,
#    }

#print("Salvataggio completato: filtered.root")
```
